```python
# Fitting straight line
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
# Data generation
# True parameters
a = 2.0
b = 1.0
```

y=a*x+b

```python
# Generate some linear data with noise
x = torch.linspace(0, 1, 100)
y = a * x + b + torch.randn(x.size()) * 0.1
```

```python
# Plot the generated data
plt.scatter(x.numpy(), y.numpy())
plt.title('Generated Data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Model definition
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super(LinearRegressionModel, self).__init__()
        self.linear = nn.Linear(1, 1)  # One input and one output

    def forward(self, x):
        return self.linear(x.unsqueeze(1))  # x needs to be 2D for nn.Linear
```

```python
# Instantiate the model, loss function, and optimizer
model = LinearRegressionModel()
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)
```

```python
# Training loop
num_epochs = 100
for epoch in range(num_epochs):
    # Forward pass: Compute predicted y by passing x to the model
    pred_y = model(x)

    # Compute and print loss
    loss = criterion(pred_y, y.unsqueeze(1))
    if epoch % 10 == 0:
        print(f'Epoch {epoch+1}/{num_epochs}, Loss: {loss.item()}')

    # Zero gradients, perform a backward pass, and update the weights.
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
# Extract learned parameters
learned_a = model.linear.weight.item()
learned_b = model.linear.bias.item()
```

```python

# Plot the original data and the fitted line
plt.scatter(x.numpy(), y.numpy(), label='Original data')
plt.plot(x.numpy(), (learned_a * x + learned_b).detach().numpy(), label='Fitted line', color='red')
plt.title('Linear Regression Result')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
